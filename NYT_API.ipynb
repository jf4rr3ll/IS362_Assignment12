```python
import pandas as pd
import json
import urllib.request
from pandas.io.json import json_normalize
```

Next, we're going to use the API key we got to read the JSON data for the Most Shared articles on Art. 

```python
api_key = 'f3f65373e71a456f84a14cf3ec2ec33f'
url = 'https://api.nytimes.com/svc/mostpopular/v2/mostshared/Arts/30.json?'
string = url+'&api-key='+api_key

response = str(urllib.request.urlopen(string).read(), 'utf-8')
```

Now we'll put this information into a data frame:

```python
mostshared_arts = pd.read_json(response)
mostshared_arts
```

Unfortunately, the resulting data frame is not particularly useful to us. We need JSON to examine the results and give us data with more information:

```python
data = json.loads(response)
most_shared = json_normalize(data['results'])
most_shared.sort_values(by=["total_shares"], ascending=False)
```

Now we have a data frame that is much easier to read and use. We can see the abstract, the author, and information about the content of the article.
